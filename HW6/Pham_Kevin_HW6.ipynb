## Problem 1

Write a function called `group_means` that takes its input a Pandas DataFrame with exactly the columns: "labels" and "weights". Group entries by labels and return the average weights for each label. Make sure to name the column in the returned data frame "average weight" and use the "labels" as the index, but don't name the index column. 

**Example:** See an example by executing the code in the second cell below.

```python
# COMPLETE
import pandas as pd
import numpy as np
from IPython.display import display

# Return average weight for each label i.e. df[label] = avg. weight
# name of the column in the returned df 'average weight' with labels as index, DONT NAME THE INDEX COL.
# GROUP EACH LABEL and keep their attached 'weight' for each affiliated label

def group_means(X): # take input as a proper dataframe
    # YOUR CODE HERE
    
    ### GET COMPONENTS FOR NEW DATAFRAME ###
    # get just the labels for new dataframe, This is for the new index it will fall under
    group_labels = set(X.labels) # caution: This is assuming that the ordered averages will be in order
    
    # grab the weighted averages of each label
    average_weight = [] # empty array for weighted inputs
    X.set_index('labels',inplace=True) # Change the indices to make traversing through each label value properly
    for key,val in X.groupby(level=0): # traverse through each grouped LABEL, NOT individual units of the labels
        # retrieve the means of each grouped label, it has the same problem that it's assuming it's arranged in-order
        average_weight.append(val.weights.mean()) # append each respective mean for each label
    
    # convert back to a numpy array for convenience and runtime.
    average_weight = np.array(average_weight)
    
    # create and return the new dataframe with the correct adjusted indices
    X_new = pd.DataFrame({'average weight' : average_weight}, index = group_labels)
    
    return X_new

    # raise NotImplementedError()
```

The two files "WorldBank_Metadata_Countries.csv" and "WorldBank_Population.csv" are available on Canvas. The population file contains information downloaded from the world bank data portal on population counts in a number of countries from 1960 to 2020. The metadata file contains information on the world-region the country is in, and on the general income group of the country. 

Open both files in Excel and familiarize yourself with the layout of the data. Then, use Pandas to load both files in Python. Write Pandas code to create the following table:

- Include only the countries in the region "Europe & Central Asia"
- Use the "Country Name" as the row identifier or Index value.
- Include the following columns for each country:
    * population count in 2000 
    * population count in 2020 
    * percent growth in population in \% between 2000 and 2020 (this should be a negative number if the population in 2020 is smaller than in 2000). Example: if population grows from 100k to 105k, then the growth is 5\%.
    * income group of the country. List *only* the income group (e.g., "Upper middle", or "High") That is, remove the OECD information and the word "income" from the labels. 
    
Display your table in Data Frame format in your notebook so that the grader can see it. 

```python
# COMPLETE 
# YOUR CODE HERE
# open appropriate excel files
world_countries = pd.read_csv("WorldBank_MetaData_Countries.csv")
world_population = pd.read_csv("WorldBank_Population.csv")

# Include only the countries in the region "Europe & Central Asia"
Euro_CAsia = pd.DataFrame(world_countries[world_countries['Region'] == 'Europe & Central Asia'])

# Use the "Country Name" as the row identifier or Index value. This is held off until later on
Euro_CAsia.set_index('Country Code', inplace = True) # note: they are now in order and helps us find population according to their 'Country Code'


# Include the following columns for each country:

# Population count for 2000 and 2020 column
# Retrieve the population count by country from the population data
count_2k = pd.DataFrame({'Country':world_population['Country Name'],'Country Code':world_population['Country Code'], '2k count':world_population['2000'], '2020 count':world_population['2020']})
count_2k.set_index('Country Code', inplace = True) # set its index to the 'Country Code' to help slice only relevant data
#print(list(Euro_CAsia.index))

# Merge the 2000 population column to the Dataframe
Euro_CAsia = Euro_CAsia.merge(count_2k['2k count'][Euro_CAsia.index], left_index=True, right_index=True)

# Population count for 2020 column
# Merge the 2020 population column to the Dataframe
Euro_CAsia = Euro_CAsia.merge(count_2k['2020 count'][Euro_CAsia.index], left_index = True, right_index = True)


# percent growth between 2000 to 2020
diff = [] # Empty list for the difference between the populations
inc_filter = [] # Empty list to queue in just the indices of entries that do NOT have OECD in existing entries
for i in range(len(Euro_CAsia.index)): # Traverse to find difference between 2020 vs. 2000 and find entries that do NOT have OECD
    inc_filter.append(Euro_CAsia.index[i]) # retrieve the index of the good entry
    diff.append(Euro_CAsia['2020 count'][i] - Euro_CAsia['2k count'][i]) # find the population differences
        
Euro_CAsia = Euro_CAsia.loc[inc_filter] # apply the filter
# remove all OCED info and 'income' in the column
Euro_CAsia['Income Group'] = Euro_CAsia['Income Group'].str.replace("nonOECD","").str.replace("OECD","").str.replace("income","")


diff_p = pd.Series(diff, index = Euro_CAsia.index) # create a series with just the percent differences
# Series needs a name apparently?
Euro_CAsia = Euro_CAsia.merge(diff_p.rename("Percent Growth"), left_index = True, right_index = True) # merge it with existing dataframe
Euro_CAsia["Percent Growth"] = round((Euro_CAsia["Percent Growth"]/Euro_CAsia["2k count"])*100,3) # convert the percent growth 

# Change the indices to the country name instead.
Euro_CAsia.set_index('Country Name', inplace = True)
display(Euro_CAsia) # display completed table

# raise NotImplementedError()
```
