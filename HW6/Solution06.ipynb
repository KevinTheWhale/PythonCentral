## Problem 1

Write a function called `group_means` that takes its input a Pandas DataFrame with exactly the columns: "labels" and "weights". Group entries by labels and return the average weights for each label. Make sure to name the column in the returned data frame "average weight" and use the "labels" as the index, but don't name the index column. 

**Example:** See an example by executing the code in the second cell below.

```python
import pandas as pd
import numpy as np
from IPython.display import display

def group_means(X):
    X_out = X.groupby('labels').mean().rename(columns={'weights':'average weight'})
        # group by the labels, compute the group means and get rid of the name of the index column
    X_out.index.name = None
    return X_out
```

- Include only the countries in the region "Europe & Central Asia"
- Use the "Country Name" as the row identifier or Index value.
- Include the following columns for each country:
    * population count in 2000 
    * population count in 2020 
    * percent growth in population in \% between 2000 and 2020 (this should be a negative number if the population in 2020 is smaller than in 2000). Example: if population grows from 100k to 105k, then the growth is 5\%.
    * income group of the country. List *only* the income group (e.g., "Upper middle", or "High") That is, remove the OECD information and the word "income" from the labels. 
    
Display your table in Data Frame format in your notebook so that the grader can see it. 

```python
import pandas as pd

# obviously, you would be using a different path on your machine
file1 = "/Users/mbremer/SJSU/Math167PS/Data/HomeworkData/WorldBank_Population.csv"
file2 = "/Users/mbremer/SJSU/Math167PS/Data/HomeworkData/WorldBank_Metadata_Countries.csv"

population = pd.read_csv(file1) # read in files
meta = pd.read_csv(file2)

meta = meta[meta.Region=='Europe & Central Asia'] # subset the meta file to keep only Europe and Central Asia

Table = pd.merge(population, meta, on = 'Country Code') # merge both files
Table['Income Group'] = [s.replace(': OECD','').replace(': nonOECD','').replace(' income','') for s in Table['Income Group']]
Table = Table[['Country Name_x', '2000', '2020', 'Income Group']] # subset columns
Table.rename(columns={'Country Name_x': 'Country Name', '2000':'pop 2000', '2020':'pop 2020'}, inplace = True)
Table['pop growth']=(Table['pop 2020']/Table['pop 2000']-1)*100 # compute population growth (in %)
Table.set_index('Country Name', inplace = True) # set country name as index
Table = Table[['pop 2000', 'pop 2020', 'pop growth', 'Income Group']] # reorder columns
Table
```
